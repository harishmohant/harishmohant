```python
from scipy import misc
ascent_images = misc.ascent()
```

```python
import matplotlib.pyplot as plt
plt.grid(False)
plt.gray()
plt.axis('off')
plt.imshow(ascent_images)
plt.show()
```

```python
import numpy as np
image_transformed = np.copy(ascent_images)
size_x = image_transformed.shape[0]
size_y = image_transformed.shape[1]
```

```python
filter =[ [0,1,0], [1,-4,1],[0,1,0]]
weight = 1
```

```python
for x in range(1,size_x-1):
  for y in range(1,size_y-1):
    convolution = 0.0
    convolution = convolution + (ascent_images [x-1,y-1]*filter[0][0])
    convolution = convolution + (ascent_images [x-1,y]*filter[0][1])
    convolution = convolution + (ascent_images [x-1,y+1]*filter[0][2])
    convolution = convolution + (ascent_images [x,y-1]*filter[1][0])
    convolution = convolution + (ascent_images [x,y]*filter[1][1])
    convolution = convolution + (ascent_images [x,y+1]*filter[1][2])
    convolution = convolution + (ascent_images [x+1,y-1]*filter[2][0])
    convolution = convolution + (ascent_images [x+1,y]*filter[2][1])
    convolution = convolution + (ascent_images [x+1,y+1]*filter[2][2])
    convolution = convolution * weight

    if( convolution < 0):
      convolution = 0
    if (convolution >255):
      convolution = 255
      
    image_transformed[x,y] = convolution 
```

```python
plt.gray()
plt.grid(False)
plt.imshow(image_transformed)
plt.show()
```
